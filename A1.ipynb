## Question 1

Compare the mean prices of Regular Unleaded Gasoline among different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") in 2023, and plot a corresponding bar chart. Please ensure that the bars are sorted by the price amounts.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('energy-price.csv')

# Replace date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filter data for the year 2023
df_2023 = df[df['Date'].dt.year == 2023]

# Filter data for Regular Unleaded Gasoline
df_gasoline = df_2023[df_2023['Fuel Type'] == 'Regular Unleaded Gasoline']

# Select columns to compare
cities = ['Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor', 'London', 'Peterborough',
          'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound', "St. Catharine's"]
columns_to_select = ['Date'] + cities
df_gasoline = df_gasoline[columns_to_select]

# Calculate mean prices for each city
mean_prices = df_gasoline.mean().drop('Date').sort_values()

# Plot bar chart
plt.figure(figsize=(13, 5))
mean_prices.plot(kind='bar', color='lightgreen')
plt.title('Mean Prices of Regular Unleaded Gasoline in Different Cities (2023)')
plt.xlabel('Cities')
plt.ylabel('Mean Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Question 3

Plot a histogram of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023, using 15 bins.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('energy-price.csv')

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filter data for the year 2023 and Toronto West/Ouest
df_toronto_west_2023 = df[(df['Date'].dt.year == 2023) & (df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Toronto West/Ouest'] != 0)]

# Filter data to Regular Unleaded Gasoline prices for Toronto West/Ouest
gasoline_prices_toronto_west = df_toronto_west_2023['Toronto West/Ouest']

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(gasoline_prices_toronto_west, bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of Regular Unleaded Gasoline Prices in Toronto West/Ouest (2023)')
plt.xlabel('Gasoline Prices')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Question 4

Plot the normalized ECDF (Empirical Cumulative Distribution Function) of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('energy-price.csv')

# Change Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract data for the year 2023 and Toronto West/Ouest
df_toronto_west_2023 = df[(df['Date'].dt.year == 2023) & (df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Toronto West/Ouest'] != 0)]

# Filter Regular Unleaded Gasoline prices for Toronto West/Ouest
gasoline_prices_toronto_west = df_toronto_west_2023['Toronto West/Ouest']

# Calculate ECDF
x = np.sort(gasoline_prices_toronto_west)
y = np.arange(1, len(x) + 1) / len(x)

# Plot normalized ECDF
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='none', linestyle='-', drawstyle='steps-post', color='grey') 
plt.title('ECDF of Regular Unleaded Gasoline Prices in Toronto West/Ouest (2023)')
plt.xlabel('Gasoline Prices')
plt.ylabel('ECDF')
plt.margins(0.02)
plt.tight_layout()
plt.show()
```

## Question 5

Plot a boxplot to display the distribution of Regular Unleaded Gasoline prices across different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") from 2021 onwards (including 2021).


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('energy-price.csv')

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filter data from 2021 onwards
df_2021_onwards = df[df['Date'].dt.year >= 2021]

# Filter data to Regular Unleaded Gasoline
df_gasoline = df_2021_onwards[df_2021_onwards['Fuel Type'] == 'Regular Unleaded Gasoline']

# Select columns
cities = ['Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor', 'London', 'Peterborough',
          'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound', "St. Catharine's"]
columns_to_select = cities + ['Date', 'Fuel Type']
df_gasoline_cities = df_gasoline[columns_to_select]

# Plot boxplot
plt.figure(figsize=(13, 6))
df_gasoline_cities.boxplot(column=cities, grid=False)
plt.title('Boxplot of Regular Unleaded Gasoline Prices Across Cities (2021 Onwards)')
plt.xlabel('City')
plt.ylabel('Gasoline Prices')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
